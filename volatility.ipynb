```python
import datetime as dt
from dateutil.relativedelta import relativedelta
import requests
import pandas as pd
import math

#Ticker
ticker = "lulu".upper()

#Date Calculations
today = dt.datetime.today()
print(today.strftime('%Y-%m-%d'))

today_minus_90days = today - relativedelta(days=90)
print(today_minus_90days.strftime('%Y-%m-%d'))

today = today.strftime('%Y-%m-%d')
today_minus_90days = today_minus_90days.strftime('%Y-%m-%d')
```

```python
polygon_api_key = open("polygon_io_api_key.txt").read()

#Pull the last 90 days of daily stock data
stock_url = f'https://api.polygon.io/v2/aggs/ticker/AAPL/range/1/day/{today_minus_90days}/{today}?adjusted=true&sort=asc&limit=120&apiKey={polygon_api_key}'
stock_r = requests.get(stock_url)
stock_data = stock_r.json()
print(stock_data)

#Turn daily
stock_df = pd.DataFrame(stock_data['results'])

#convert from unixtime to date
stock_df['t'] = stock_df['t'].apply(lambda t:dt.datetime.fromtimestamp(t/1000).strftime('%Y-%m-%d'))

vwap_average = stock_df['vw'].sum()/len(stock_df['vw'])
stock_df['vw_difference'] = stock_df['vw'] - vwap_average
```

```python
vwap_average = stock_df['vw'].sum()/len(stock_df['vw'])
stock_df['vw_difference'] = stock_df['vw'] - vwap_average
stock_df['vw_square'] = stock_df['vw_difference'] ** 2
print(math.sqrt((stock_df['vw_square'].sum()/len(stock_df['vw']))))
print(stock_df)
```
